```python
# Initialize the interface with your desired settings
mg_interface = interface.MetallogenInterface(
    working_directory="./scratch_dir",
    save_directory="./my_results",
    final_relax=True,
    num_conformer=1,
    calculator_type="xtb_s" # options: "xtb", "orca", "gaussian", "xtb_gaussian"
)

# Run via MSMILES string
smiles_str = "[Zr+4]|[Cl-:2]|[Cl-:3]|[N:1]1=C(C[C-:4]2[CH:4]=[CH:4][CH:4]=[CH:4]2)C=CC=C1(C[C-:5]3[CH:5]=[CH:5][CH:5]=[CH:5]3)|5_trigonal_bipyramidal"
results = mg_interface.generate(smiles=smiles_str)

# OR Run via SDF directory
# results = mg_interface.generate(input_directory="./path/to/sdf_folder")
```
